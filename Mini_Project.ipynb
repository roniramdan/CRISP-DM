# **DATA PREPARATION**

```python
#mengubah tipe data float menjadi int
df_smoker['cigs_per_day'] = df_smoker['cigs_per_day'].fillna(0).astype('int')
```

```python
#mengubah tipe data float menjadi int
df_smoker['chol'] = df_smoker['chol'].fillna(0).astype('int')
```

```python
df_smoker.dropna(how='any')
```

```python
# Menghapus baris-baris yang memenuhi kriteria dia ngerokok tapi jumlah batang rokok nya nol
df_smoker = df_smoker.drop(df_smoker[(df_smoker['current_smoker'] == 'yes') & (df_smoker['cigs_per_day'] == 0)].index)

# Menampilkan DataFrame setelah baris-baris yang dipilih dihapus
print(df_smoker)
```

```python
#replace inkonsistensi 'f', 'm' ke 'female', 'male' -> data preparation
df_smoker['sex'] = df_smoker['sex'].replace({'f': 'female', 'm': 'male'})

print(df_smoker.sex.unique())
```

```python
index_to_drop = df_smoker[df_smoker['age'] == 1000].index

# Menghapus baris dengan indeks yang ditemukan
df_smoker = df_smoker.drop(index_to_drop)
```

```python
index_to_drop = df_smoker[df_smoker['age'] == 150].index

# Menghapus baris dengan indeks yang ditemukan
df_smoker = df_smoker.drop(index_to_drop)
```

```python
index_to_drop = df_smoker[df_smoker['heart_rate'] == -1].index

# Menghapus baris dengan indeks yang ditemukan
df_smoker = df_smoker.drop(index_to_drop)
```

```python
index_to_drop = df_smoker[df_smoker['heart_rate'] == 0].index

# Menghapus baris dengan indeks yang ditemukan
df_smoker = df_smoker.drop(index_to_drop)
```

```python
# Filter baris yang memiliki nilai 'chol' di atas 350
indexes_to_drop = df_smoker[df_smoker['chol'] > 350].index

# Hapus baris dengan index yang memenuhi kriteria
df_smoker.drop(indexes_to_drop, inplace=True)

# Cetak DataFrame setelah penghapusan
print(df_smoker)
```

```python
# Filter baris yang memiliki nilai 'chol' di bawah 120
indexes_to_drop = df_smoker[df_smoker['chol'] < 120].index

# Hapus baris dengan index yang memenuhi kriteria
df_smoker.drop(indexes_to_drop, inplace=True)

# Cetak DataFrame setelah penghapusan
print(df_smoker)
```

# **MODELING EDA**

**EDA Menghitung rata-rata kolesterol berdasarkan jenis kelamin dan status perokok 'yes'**



```python
# Lakukan filter pada DataFrame untuk hanya mempertimbangkan data yang memiliki nilai 'current_smoker' sama dengan 'yes'
smoking_yes = df_smoker[df_smoker['current_smoker'] == 'yes']

# Menghitung rata-rata kolesterol berdasarkan jenis kelamin dan status perokok 'yes'
mean_chol_by_gender_smoker_yes = smoking_yes.groupby('sex')['chol'].mean()

# Membuat visualisasi bar chart
plt.figure(figsize=(8, 6))
mean_chol_by_gender_smoker_yes.plot(kind='bar', color=['pink', 'blue'])
plt.title('Rata-rata Kolesterol untuk Perokok (Status: Yes) Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Rata-rata Kolesterol')
plt.xticks(rotation=0)
plt.show()
```

```python
# Lakukan filter pada DataFrame untuk hanya mempertimbangkan data yang memiliki nilai 'current_smoker' sama dengan 'yes'
smoking_yes = df_smoker[df_smoker['current_smoker'] == 'yes']

# Menghitung rata-rata kolesterol berdasarkan jenis kelamin dan status perokok 'yes'
mean_chol_by_gender_smoker_yes = smoking_yes.groupby('sex')['chol'].mean()

# Cetak rata-rata kolesterol berdasarkan jenis kelamin dan status perokok 'yes'
print("Rata-rata kolesterol berdasarkan jenis kelamin dan status perokok 'yes':")
print(mean_chol_by_gender_smoker_yes)
```

**EDA Menghitung rata-rata tingkat kolesterol untuk lelaki yang merokok dan yang tidak merokok**

```python
male_smokers_df = df_smoker[(df_smoker['sex'] == 'male') & (df_smoker['current_smoker'] == 'yes')]

# Memfilter data untuk lelaki yang tidak merokok
male_non_smokers_df = df_smoker[(df_smoker['sex'] == 'male') & (df_smoker['current_smoker'] == 'no')]

# Menghitung rata-rata tingkat kolesterol untuk lelaki yang merokok dan yang tidak merokok
mean_chol_smokers = male_smokers_df['chol'].mean()
mean_chol_non_smokers = male_non_smokers_df['chol'].mean()

# Plot histogram dengan garis vertikal menunjukkan rata-rata tingkat kolesterol
plt.figure(figsize=(10, 6))
sns.histplot(data=male_smokers_df, x='chol', color='blue', label=f'Male Smokers (Mean: {mean_chol_smokers:.2f})', kde=True)
sns.histplot(data=male_non_smokers_df, x='chol', color='green', label=f'Male Non-Smokers (Mean: {mean_chol_non_smokers:.2f})', kde=True)

plt.axvline(x=mean_chol_smokers, color='blue', linestyle='--', linewidth=1.5)
plt.axvline(x=mean_chol_non_smokers, color='green', linestyle='--', linewidth=1.5)

plt.xlabel('Tingkat Kolesterol')
plt.ylabel('Frekuensi')
plt.title('Histogram Tingkat Kolesterol Rata-rata Lelaki Merokok dan Tidak Merokok')
plt.legend()

plt.show()
```

**EDA Menghitung rata-rata detak jantung untuk lelaki yang merokok dan yang tidak merokok**

```python

# Memfilter data untuk lelaki yang merokok
male_smokers_df = df_smoker[(df_smoker['sex'] == 'male') & (df_smoker['current_smoker'] == 'yes')]

# Memfilter data untuk lelaki yang tidak merokok
male_non_smokers_df = df_smoker[(df_smoker['sex'] == 'male') & (df_smoker['current_smoker'] == 'no')]

# Menghitung rata-rata detak jantung untuk lelaki yang merokok dan yang tidak merokok
mean_hr_smokers = male_smokers_df['heart_rate'].mean()
mean_hr_non_smokers = male_non_smokers_df['heart_rate'].mean()

# Plot histogram untuk detak jantung lelaki yang merokok dan tidak merokok
plt.figure(figsize=(10, 6))
sns.histplot(data=male_smokers_df, x='heart_rate', color='blue', label='Male Smokers (Mean: {:.2f})'.format(mean_hr_smokers), kde=True)
sns.histplot(data=male_non_smokers_df, x='heart_rate', color='red', label='Male Non-Smokers (Mean: {:.2f})'.format(mean_hr_non_smokers), kde=True)

plt.xlabel('Detak Jantung')
plt.ylabel('Frekuensi')
plt.title('Histogram Detak Jantung untuk Male Smokers dan Non-Smokers')
plt.legend()

plt.show()
```
